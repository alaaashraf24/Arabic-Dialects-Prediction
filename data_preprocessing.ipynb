```python
#Import libraries
import pandas as pd
import re
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data from the CSV file
data = pd.read_csv('dialects_data.csv')
```

```python
# Function to clean text
def clean_text(text):
    text = re.sub(r'@\w+', '', text)  # Remove mentions
    text = re.sub(r'http\S+', '', text)  # Remove URLs
    text = re.sub(r'[^\w\s]', '', text)  # Remove emojis and punctuation
    text = re.sub(r'\d+', '', text)  # Remove digits
    text = re.sub(r'[a-zA-Z]', '', text)  # Remove English characters
    text = re.sub(r'\n', ' ', text)  # Remove newlines
    text = text.strip()
    return text
```

```python
# Apply text cleaning
data['cleaned_text'] = data['text'].apply(clean_text)
```

```python
data
```

```python
# Encode the labels
label_encoder = LabelEncoder()
data['label'] = label_encoder.fit_transform(data['dialect'])
```

```python
data
```

```python
# Save the cleaned data to a CSV file
data.to_csv('preprocessed_dialects_data.csv', index=False)
```

```python
# Plot the distribution of dialects
plt.figure(figsize=(10, 6))
sns.countplot(x=data['dialect'])
plt.title('Distribution of Dialects')
plt.xlabel('Dialect')
plt.ylabel('Count')
plt.show()
```

```python
# Word count distribution
data['word_count'] = data['cleaned_text'].apply(lambda x: len(x.split()))
plt.figure(figsize=(10, 6))
sns.histplot(data['word_count'], bins=30, kde=True)
plt.title('Word Count Distribution')
plt.xlabel('Word Count')
plt.ylabel('Frequency')
plt.show()
```
